### Ingesting to Delta Table in Refined Layer

Estamos usando o modo overwrite por enquanto para sobre escrever.
Em uma situação em produção, onde temos dado novo chegando a cada minuto, temos que usar o modo insert, update or upsert, como o "Merge".
Em produção, o certo também é criar uma tabela com schema definido, com unique_constraints properties.

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, TimestampType, DoubleType, BooleanType

schema = StructType([
    StructField("audit_ts", TimestampType(), True),
    StructField("playlist", StringType(), True),
    StructField("track_id", StringType(), True),
    StructField("track_name", StringType(), True),
    StructField("artist", StringType(), True),
    StructField("artist_id", StringType(), True),
    StructField("artist_image", DoubleType(), True),
    StructField("album", StringType(), True),
    StructField("album_image", StringType(), True),
    StructField("popularity", IntegerType(), True),
    StructField("duration_ms", IntegerType(), True),
    StructField("track_name_y", StringType(), True),
    StructField("artist_y", StringType(), True),
    StructField("album_y", StringType(), True),
    StructField("genre", StringType(), True),
    StructField("subgenre", StringType(), True),
    StructField("danceability", DoubleType(), True),
    StructField("energy", DoubleType(), True),
    StructField("mood", StringType(), True),
    StructField("release_year", IntegerType(), True),
    StructField("era", StringType(), True),
    StructField("lyrics_theme", StringType(), True),
    StructField("instrumentalness", DoubleType(), True),
    StructField("explicit", BooleanType(), True)
])

enriched_tracks = spark.createDataFrame(df_rfd_tracks, schema=schema)

enriched_tracks.write.format("delta").mode("overwrite").saveAsTable("capstone_project.capstone_project_rfd.enriched_tracks")
```
